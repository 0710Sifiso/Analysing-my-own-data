```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv('financial_risk_dataset.csv')

df.head(10)
```

```python
#Checking the shape and type of the data
print(np.shape(df))
df.dtypes
```

```python
# Cleaning up the data
# Only want Divorced, Graduate, with Credit score that is above 520
```

```python
df[df['Marital_Status'] == 'Divorced']
```

```python
df[(df['Education_Level'] == 'Graduate') & 
   (df['Marital_Status'] == 'Divorced')]
```

```python
df2 = df[(df['Education_Level'] == 'Graduate') & 
   (df['Marital_Status'] == 'Divorced') & 
   (df['Credit_Score'] <= 540)]
```

```python
df2.head()
```

```python
#Drop columns that we do not need: Loan_ID, Property_Area, Probability_of_Default, Default_Status
```

```python
df2 = df2.drop(['Loan_ID', 'Property_Area', 'Probability_of_Default', 'Default_Status'], axis = 1)
```

```python
df2.head()
```

```python
#removing null values
df2.isna().sum()
```

```python
df2=df2.dropna()
df2.shape
df2.isna().sum()
```

```python
#Checking for duplicates
df2[df2.duplicated()==True]
```

```python
df2.reset_index(drop =True)
```

```python
df2.dtypes
```

```python
# Select only the numerical columns
numeric_columns = df2.select_dtypes(include=['number'])

# Calculate the average for each numerical column
average_values = numeric_columns.mean()
print(average_values)

```

```python
#Creating plots
```

```python
sns.histplot(df2['Applicant_Age'])
```

```python
#Credit Score vs. Age (Line Plot)
plt.figure(figsize=(10, 5))
sns.lineplot(x=df2['Applicant_Age'], y=df2['Credit_Score'], marker='o', color='b')
plt.xlabel('Applicant Age')
plt.ylabel('Credit Score')
plt.title('Credit Score vs. Age')
plt.grid(True)
plt.show()
```

```python
# Income vs. Debt-to-Income Ratio (Scatter Plot)
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df2['Applicant_Income'], y=df2['Debt_to_Income_Ratio'], alpha=0.6, color='r')
plt.xlabel('Applicant Income')
plt.ylabel('Debt-to-Income Ratio')
plt.title('Income vs. Debt-to-Income Ratio')
plt.show()
```

```python
# Loan Amount vs. Interest Rate 
plt.figure(figsize=(12, 6))

sns.lineplot(x=df2['Interest_Rate'], y=df2['Loan_Amount'], palette='coolwarm')

plt.xlabel('Interest Rate (%)')
plt.ylabel('Loan Amount')
plt.title('Loan Amount vs. Interest Rate')
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=df2['Years_in_Employment'], y=df2['Loan_Amount'], color='g', alpha=0.7)
plt.xlabel('Years in Employment')
plt.ylabel('Loan Amount')
plt.title('Years in Employment vs. Loan Amount')
plt.grid(True)
plt.show()

```

#The average age of the applicants is around 42.7 years, which suggests a relatively mature group with potentially more stable financial backgrounds.

They have an average of 5.22 years in employment, which indicates a moderate to long-term job tenure, contributing to a sense of job stability.

The applicants have an average of 1.06 dependents, suggesting a moderate financial responsibility due to family commitments.

The combined income (applicant and coapplicant) stands at an average of $62,430 and $29,084, respectively, which suggests that the applicants have a reasonable income base.

With an average credit score of 488.7, it's worth noting that many of these applicants fall below the desirable threshold of 520. However, given the credit score constraint, a more specific sub-set of applicants with scores above 520 could represent a more financially reliable group.

Existing debt averages at $51,301, which is a noteworthy liability that could affect loan eligibility.

The loan amount requested is an average of $37,111, indicating a moderate loan application relative to the applicants' financial backgrounds.

The average collateral value is $77,674, which provides substantial security for the loan.

With payment delays averaging 1.44 in the past 6 months and a debt-to-income ratio of 0.67, this suggests some financial strain but is somewhat mitigated by the collateral backing.

In conclusion, the applicants, on average, present a moderately risky financial profile, primarily due to their existing debt and relatively low credit scores. However, the collateral value and loan term structure could reduce some of this risk. Further filtering to include only those with credit scores above 520 would likely enhance the financial reliability of this group.
